```python
import numpy as np
```

```python
batch_size = 32
in_channel = 3
out_channel = 4
input_size = 100
kernel_size = 5
stride = 2
```

```python
seed = 123
np.random.seed(seed)
x = np.random.rand(batch_size, in_channel, input_size)
W = np.random.rand(out_channel, in_channel, kernel_size)
b = np.random.rand(out_channel)
print(x.shape)
print(W.shape)
print(b.shape)
```

```python
batch_size = x.shape[0]
in_width = x.shape[-1]
out_width = int(np.floor(1+(in_width-kernel_size)/stride))
out = np.zeros((batch_size,out_channel,out_width))
for n in range(batch_size):
    for k2 in range(out_channel):
        bias = b[k2]
        for t in range(out_width):
            affine = 0
            s = stride*t
            for k1 in range(in_channel):     
                affine += np.dot(W[k2][k1],x[n][k1][s:s+kernel_size])
            out[n][k2][t] = bias + affine
print(out.shape)
out[0]
```
