```python
l1 = [0, 0, 1, 1]
l2 = [0, 1, 0, 1]
O = [1, 1, 1, 0]

W1 = 0
W2 = 0
b = 0

num_epoch = 100
eta = 0.01

W1list = [W1]
W2list = [W2]
blist = [b]

def phi(num):
    return 0 if num<=0 else 1

for iteration in range(num_epoch):
    if iteration%2 !=0 :
        for i in range(4):
            t = phi(W1 * l1[i] + W2 * l2[i] + b)
            y = O[i]
#             print(y-t)
            W1 += eta*l1[i]*(y-t)
            W2 += eta*l2[i]*(y-t)
            b += eta*(y-t)
    else:
        for j in range(4):
            i = 3-j
            t = phi(W1 * l1[i] + W2 * l2[i] + b)
            y = O[i]
#             print(y-t)
            W1 += eta*l1[i]*(y-t)
            W2 += eta*l2[i]*(y-t)
            b += eta*(y-t)
    W1list.append(W1)
    W2list.append(W2)
    blist.append(b)

print(phi(W1*0 + W2*0 + b))
print(phi(W1*1 + W2*0 + b))
print(phi(W1*0 + W2*1 + b))
print(phi(W1*1 + W2*1 + b))
print(W1)
print(W2)
print(b)
```
