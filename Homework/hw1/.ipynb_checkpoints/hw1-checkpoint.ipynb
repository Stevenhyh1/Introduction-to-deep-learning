```python
x = np.arange(80).reshape((8,10))
y = np.zeros((8,10))
y[:,4] = 1

exps = np.exp(x - np.max(x, axis = 1,keepdims = True))
softmax = exps/np.sum(exps,axis =1,keepdims = True)
loss = -np.sum(y*softmax, axis = 1)
```

```python
mean = np.mean(x, axis=0,keepdims = True)
var = np.var(x,axis=0,keepdims = True)
print(mean.shape)
print(var.shape)
```

```python
eps = 1e-8
norm = (x-mean)/(var+eps)
print(norm.shape)
```

```python
gamma = np.ones((1,10))
beta = np.ones((1,10))
output = norm*gamma + beta
print(output.shape)
```

```python
seed = 123
np.random.seed(seed)
delta = np.random.randn(8,10)
```

```python
dnorm = delta*gamma
print(dnorm)
```

```python
dbeta = np.sum(delta, axis = 0, keepdims = True)
dgamma = np.sum(delta * norm, axis = 0, keepdims = True)
print(dbeta.shape, dgamma.shape)
```

```python
dvar = -0.5*np.sum((x-mean)*np.power((var + eps),-3/2), axis = 0, keepdims = True)
print(dvar.shape)
```

```python
dmean = -np.sum(dnorm*np.power((var + eps),-0.5), axis = 0, keepdims = True) - 2*dvar*np.mean(x-mean, axis = 0, keepdims = True)
print(dmean.shape)
```

```python
m = len(x)
dx = dnorm*np.power((var+eps),-0.5) + dvar*(2/m*(x-mean)) + dmean/m
print(dx.shape)
```
